This time series visualization represents three key variables in Sri Lanka’s tea industry over time:

Nett Sale Average (Rs) – in red

Low Grown Average (Rs) – in blue

Bought Leaf Rate (Rs) – in green

Let’s break down the graph and relate it to real-world industry insights:

### 1. Overall Trends (1998–2023)
Nett Sale Average and Low Grown Average show similar trends, closely moving together.

Bought Leaf Rate is consistently much lower, as expected, since it’s the price paid to smallholders/farmers for green leaf—not for processed tea.

### 2. Early Years (~1998 to 2008)
Modest and relatively stable increases in prices.

This reflects a period of moderate global tea demand, steady supply, and limited inflationary pressure.

The civil war in Sri Lanka (ended in 2009) possibly affected infrastructure and logistics, contributing to slower industry growth.

### 3. 2009–2014: Post-war Growth & Global Commodity Boom
A noticeable upward trend post-2009: Reflects the post-war economic recovery.

Increased investment in the plantation sector, improved logistics, and growing exports likely contributed.

Global commodity prices were also rising during this time.

### 4. 2015–2019: Volatility, Policy Shifts, and Market Fluctuations
Fluctuations in all three curves, with periodic peaks and drops.

Factors:

Weather variability (e.g., droughts, floods).

Ban on glyphosate (2015–2018)—affected crop yield and quality.

Export demand fluctuations, especially from key markets like Iran, Iraq, and Russia.

Currency depreciation—affected input costs and export earnings.

### 5. 2020–2022: COVID-19 Pandemic and Economic Crisis
Sharp surge in Nett Sale Average and Low Grown Average (~2021–2022), peaking in 2022.

Likely reasons:

Global supply chain disruptions pushed prices up.

Sri Lanka’s economic crisis (2021–2022): Rupee devaluation made exports more profitable in LKR.

Fertilizer import ban (2021) reduced yields, limiting supply and pushing prices up.

### 6. 2023 Onwards: Normalization and Decline
After 2022 peak, prices drop sharply.

This could be due to:

Recovery in global supply chains.

Slight stabilization of Sri Lankan economy.

Increased competition from other tea-exporting countries.

Possible demand reduction due to global inflation and recession fears.

### 7. Bought Leaf Rate Dynamics
Remains consistently low but lags behind sale averages.

Notably, it also increased in 2022, but to a lesser extent.

Reflects the persistent income inequality between estate factories and smallholder farmers.

A major industry challenge: ensuring fair value transmission to producers.

### Key Notes:
Market volatility in the Sri Lankan tea sector is heavily influenced by global demand, domestic policies, climate, and economic crises.

The 2022 spike is an outlier, driven by local economic collapse and global supply shocks.

There is systemic undercompensation for bought leaf suppliers compared to sale averages.

Strategic reforms are needed for price stabilization, value chain equity, and export competitiveness.
